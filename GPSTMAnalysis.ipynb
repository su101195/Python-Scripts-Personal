```python
import pandas as pd
import numpy as np
from mda_core.time import Time
import os
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
```

# GPS 1 & GPS 2 TM Analysis

## Load the data

```python
df_gps1 = pd.DataFrame()
df_gps2 = pd.DataFrame()
df_gps = pd.DataFrame()

filepath = f'../data/GPS_Module_Test_Full_Dump.csv'

df_gps = pd.read_csv(filepath, sep = '\t')
df_gps = df_gps.fillna(0)
df_gps.columns
```

```python
df_gps
```

## Pre-processing the GPS TM data

```python
df_gps1 = df_gps
gps1Time = []
columns_to_keep = ['gpsPEcef_PosX', 'gpsPEcef_PosY', 'gpsPEcef_PosZ', 'gpsPEcef_VelX', 'gpsPEcef_VelY', 'gpsPEcef_VelZ', 'gpsPUtc_Year',
       'gpsPUtc_Month', 'gpsPUtc_Day', 'gpsPUtc_Hour', 'gpsPUtc_Min','gpsPUtc_MilliSec']
for col in df_gps1.columns:
    if col not in columns_to_keep:
        df_gps1 = df_gps1.drop(col,axis=1)

df_gps1 = df_gps1[(df_gps1['gpsPEcef_PosX'] != 0.0) & (df_gps1['gpsPUtc_Year'] == 2024)]
for id, row in df_gps1.iterrows():
    gps1Time.append(datetime(
    year = int(row['gpsPUtc_Year']),
    month = int(row['gpsPUtc_Month']),
    day = int(row['gpsPUtc_Day']),
    hour = int(row['gpsPUtc_Hour']),
    minute = int(row['gpsPUtc_Min']),
    second = int(row['gpsPUtc_MilliSec']/1e3)))
df_gps1['gpsUtc_Time'] = gps1Time
# df_gps1 = df_gps1.rename(columns={'gpsPUtc_MilliSec': 'gpsPUtc_Sec'})
# df_gps1['gpsPUtc_Sec'] = df_gps1['gpsPUtc_Sec']/1e3
df_gps1 = df_gps1.drop_duplicates()
df_gps1 = df_gps1.reset_index(drop=True)
df_gps1
```

```python
df_gps2 = df_gps
gps2Time = []
columns_to_keep = ['gpsSEcef_PosX', 'gpsSEcef_PosY', 'gpsSEcef_PosZ', 'gpsSEcef_VelX', 'gpsSEcef_VelY', 'gpsSEcef_VelZ', 'gpsS_Year',
       'gpsS_Month', 'gpsS_Day', 'gpsS_Hour', 'gpsS_Min','gpsS_Sec']
for col in df_gps2.columns:
    if col not in columns_to_keep:
        df_gps2 = df_gps2.drop(col,axis=1)

df_gps2 = df_gps2[(df_gps2['gpsSEcef_PosX'] != 0.0) & (df_gps2['gpsS_Year'] == 2024)]
for id, row in df_gps2.iterrows():
    gps2Time.append(datetime(
    year = int(row['gpsS_Year']),
    month = int(row['gpsS_Month']),
    day = int(row['gpsS_Day']),
    hour = int(row['gpsS_Hour']),
    minute = int(row['gpsS_Min']),
    second = int(row['gpsS_Sec'])))
df_gps2['gpsUtc_Time'] = gps2Time
df_gps2 = df_gps2.drop_duplicates()
df_gps2 = df_gps2.reset_index(drop=True)
df_gps2
```

### Time correlated Error between GPS1 & GPS2

```python
start_specific_timestamp = pd.Timestamp('2024-09-11 12:28:14.000')
end_specific_timestamp = pd.Timestamp('2024-09-11 19:59:26.000')

error_df = pd.DataFrame()

df_gps1 = df_gps1[(df_gps1['gpsUtc_Time'] >= start_specific_timestamp) & (df_gps1['gpsUtc_Time'] <= end_specific_timestamp)]
df_gps2 = df_gps2[(df_gps2['gpsUtc_Time'] >= start_specific_timestamp) & (df_gps2['gpsUtc_Time'] <= end_specific_timestamp)]

error_df['UTC time'] = df_gps1['gpsUtc_Time']
error_df['px'] = df_gps1['gpsPEcef_PosX'] - df_gps2['gpsSEcef_PosX']
error_df['py'] = df_gps1['gpsPEcef_PosY'] - df_gps2['gpsSEcef_PosY']
error_df['pz'] = df_gps1['gpsPEcef_PosZ'] - df_gps2['gpsSEcef_PosZ']
error_df['vx'] = df_gps1['gpsPEcef_VelX'] - df_gps2['gpsSEcef_VelX']
error_df['vy'] = df_gps1['gpsPEcef_VelY'] - df_gps2['gpsSEcef_VelY']
error_df['vz'] = df_gps1['gpsPEcef_VelZ'] - df_gps2['gpsSEcef_VelZ']

```

```python
def Scatter_plot(df,i,str1,str2,str3,str4):
    fig = go.Figure()
    if i == 2:
        fig.add_trace(go.Scatter(x=df.index, y=df[str1]))
        fig.add_trace(go.Scatter(x=df.index, y=df[str2]))
    elif i == 3:
        fig.add_trace(go.Scatter(x=df.index, y=df[str1]))
        fig.add_trace(go.Scatter(x=df.index, y=df[str2]))
        fig.add_trace(go.Scatter(x=df.index, y=df[str3]))
    plot_font = dict(family="Courier New, monospace", size=18, color="RebeccaPurple")
    fig.update_layout(
        font_family="Courier New",
        font_color="black",
        title_font_family="Times New Roman",
        title_font_color="red",
        legend_title_font_color="green",
        title = str4
    )
    fig.show()
```

```python
# Receiver Position Error
Scatter_plot(error_df,3,'px','py','pz','GPS1 - GPS2 Position error (m)')
# Receiver Velocity Error
Scatter_plot(error_df,3,'vx','vy','vz','GPS1 - GPS2 Velocity error(ms)')  
```

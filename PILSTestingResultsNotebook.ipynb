```python
from mda_core.time import Time
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import os
from datetime import timedelta
```

```python
Operation = 'DETUMBLING'
```

```python
if Operation == 'DETUMBLING':

    Controller = 'MtqFailure_Bdot'

    if Controller == 'RateFeedback':
        filepath = f'../data/Detumbling_using_RateFeedbackCntlr.csv'

        df_detumbling = pd.DataFrame()

        df_detumbling = pd.read_csv(filepath, sep='\t')

        df_detumbling = df_detumbling.fillna(0)
        # df_detumbling = df_detumbling.drop(columns=['rateDampActv_Ctrlr'])
        df_detumbling = df_detumbling[(df_detumbling['gyroBdyInrtlCorr_NormalizedDerived'] != 0)]

        fig = px.line(df_detumbling, x='Time', y=['gyroBdyInrtlCorr_NormalizedDerived'], 
                title=f'{Controller} Controller Performance', hover_name='SysMode')
        
        fig1 = px.line(df_detumbling, x='Time', y=['gyroBdyInrtlCorr_OmegaX', 'gyroBdyInrtlCorr_OmegaY', 'gyroBdyInrtlCorr_OmegaZ'], 
                title=f'{Controller} Controller Performance', hover_data='SysMode')
        
        # Ensure datetime format
        df_detumbling['Time'] = pd.to_datetime(df_detumbling['Time'])
        
        # Find the first "Detumbling" timestamp
        first_detumbling_time = df_detumbling.loc[df_detumbling['SysMode'] == 'Detumbling', 'Time'].iloc[0]

        # Find the timestamp corresponding to Rates < 1 deg
        d_time = df_detumbling[df_detumbling['gyroBdyInrtlCorr_NormalizedDerived'] < 1.0]

        # Time taken for detumbling
        detumbled_time = pd.to_timedelta(d_time['Time'].iloc[0] - first_detumbling_time).total_seconds()
        
        print("Total detumbling time:", detumbled_time)

        
    elif Controller == 'Bdot':
        filepath = f'../data/Detumbling_using_BdotCtrlr.csv'

        df_detumbling = pd.DataFrame()

        df_detumbling = pd.read_csv(filepath, sep='\t')

        df_detumbling = df_detumbling.fillna(0)
        # df_detumbling = df_detumbling.drop(columns=['rateDampActv_Ctrlr'])
        df_detumbling = df_detumbling[(df_detumbling['gyroBdyInrtlCorr_NormalizedDerived'] != 0)]

        fig = px.line(df_detumbling, x='Time', y=['gyroBdyInrtlCorr_NormalizedDerived'], 
                title=f'{Controller} Controller Performance', hover_name='SysMode')
        
        fig1 = px.line(df_detumbling, x='Time', y=['gyroBdyInrtlCorr_OmegaX', 'gyroBdyInrtlCorr_OmegaY', 'gyroBdyInrtlCorr_OmegaZ'], 
                title=f'{Controller} Controller Performance', hover_data='SysMode')
        
        # Ensure datetime format
        df_detumbling['Time'] = pd.to_datetime(df_detumbling['Time'])
        
        # Find the first "Detumbling" timestamp
        first_detumbling_time = df_detumbling.loc[df_detumbling['SysMode'] == 'Detumbling', 'Time'].iloc[0]

        # Find the timestamp corresponding to Rates < 1 deg
        d_time = df_detumbling[df_detumbling['gyroBdyInrtlCorr_NormalizedDerived'] < 1.0]

        # Time taken for detumbling
        detumbled_time = pd.to_timedelta(d_time['Time'].iloc[0] - first_detumbling_time).total_seconds()
        
        print("Total detumbling time:", detumbled_time)

    elif Controller == 'MtqFailure_RateFeedback':
        filepath = f'../data/MtqXfailure_Detumbling_RateFeedbackCtrlr.csv'

        df_detumbling = pd.DataFrame()

        df_detumbling = pd.read_csv(filepath, sep='\t')

        df_detumbling = df_detumbling.fillna(0)
        # df_detumbling = df_detumbling.drop(columns=['rateDampActv_Ctrlr'])
        df_detumbling = df_detumbling[(df_detumbling['gyroBdyInrtlCorr_NormalizedDerived'] != 0)]

        fig = px.line(df_detumbling, x='Time', y=['gyroBdyInrtlCorr_NormalizedDerived'], 
                title=f'{Controller} Controller Performance', hover_name='SysMode')
        
        fig1 = px.line(df_detumbling, x='Time', y=['gyroBdyInrtlCorr_OmegaX', 'gyroBdyInrtlCorr_OmegaY', 'gyroBdyInrtlCorr_OmegaZ'], 
                title=f'{Controller} Controller Performance', hover_data='SysMode')
        
        # Ensure datetime format
        df_detumbling['Time'] = pd.to_datetime(df_detumbling['Time'])
        
        # Find the first "Detumbling" timestamp
        first_detumbling_time = df_detumbling.loc[df_detumbling['SysMode'] == 'Detumbling', 'Time'].iloc[0]

        # Find the timestamp corresponding to Rates < 1 deg
        d_time = df_detumbling[df_detumbling['gyroBdyInrtlCorr_NormalizedDerived'] < 1.0]

        # Time taken for detumbling
        detumbled_time = pd.to_timedelta(d_time['Time'].iloc[0] - first_detumbling_time).total_seconds()
        
        print("Total detumbling time:", detumbled_time)

    elif Controller == 'MtqFailure_Bdot':
        filepath = f'../data/MtqXfailure_Detumbling_BdotCtrlr.csv'

        df_detumbling = pd.DataFrame()

        df_detumbling = pd.read_csv(filepath, sep='\t')

        df_detumbling = df_detumbling.fillna(0)
        # df_detumbling = df_detumbling.drop(columns=['rateDampActv_Ctrlr'])
        df_detumbling = df_detumbling[(df_detumbling['gyroBdyInrtlCorr_NormalizedDerived'] != 0)]

        fig = px.line(df_detumbling, x='Time', y=['gyroBdyInrtlCorr_NormalizedDerived'], 
                title=f'{Controller} Controller Performance', hover_name='SysMode')
        
        fig1 = px.line(df_detumbling, x='Time', y=['gyroBdyInrtlCorr_OmegaX', 'gyroBdyInrtlCorr_OmegaY', 'gyroBdyInrtlCorr_OmegaZ'], 
                title=f'{Controller} Controller Performance', hover_data='SysMode')
        
        # Ensure datetime format
        df_detumbling['Time'] = pd.to_datetime(df_detumbling['Time'])
        
        # Find the first "Detumbling" timestamp
        first_detumbling_time = df_detumbling.loc[df_detumbling['SysMode'] == 'Detumbling', 'Time'].iloc[0]

        # Find the timestamp corresponding to Rates < 1 deg
        d_time = df_detumbling[df_detumbling['gyroBdyInrtlCorr_NormalizedDerived'] < 1.0]

        # Time taken for detumbling
        detumbled_time = pd.to_timedelta(d_time['Time'].iloc[0] - first_detumbling_time).total_seconds()
        
        print("Total detumbling time:", detumbled_time)

    else:

        raise ValueError('Incorrect Controller specified!')
        
        
        
 

df_detumbling
```
